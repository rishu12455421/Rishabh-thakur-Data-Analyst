### Objective:
The goal of this visualization task is to track and analyze customer retention behavior over time.  
By grouping customers into cohorts based on their first order month, we aim to understand loyalty trends and identify opportunities for improving customer lifetime value.

---


## (A) Data Preparation

We prepare the dataset by assigning each customer a "cohort month" based on their first order date, and extract the month of every order placed.


```python
# Create cohort_month and order_month
orders_df['cohort_month'] = orders_df.groupby('customer_id')['order_date'].transform('min').dt.to_period('M')
orders_df['order_month'] = orders_df['order_date'].dt.to_period('M')

orders_df[['customer_id', 'order_date', 'cohort_month', 'order_month']].head()
```

## (B) Cohort Table Creation

We create a cohort table that aggregates the number of unique customers per cohort and order month combination.


```python
# Group by cohort_month and order_month
cohort_data = orders_df.groupby(['cohort_month', 'order_month']).agg(n_customers=('customer_id', 'nunique')).reset_index()

# Pivot to matrix
cohort_pivot = cohort_data.pivot(index='cohort_month', columns='order_month', values='n_customers')

cohort_pivot
```

## (C) Calculate Retention Rates

We normalize the cohort table to show customer retention as a percentage of the original cohort size.


```python
# Retention rate matrix
cohort_size = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0).round(3)

retention_matrix

```

## (D) Visualize Retention Heatmap

We plot a heatmap of the retention matrix to easily visualize customer behavior across cohorts over time.


```python
# Plot the heatmap
plt.figure(figsize=(14,8))
sns.heatmap(retention_matrix, annot=True, fmt='.0%', cmap='Blues')
plt.title('Customer Retention by Cohort Month')
plt.xlabel('Order Month')
plt.ylabel('Cohort Month')
plt.show()
```

## (E) Interpretation

The cohort heatmap reveals the following insights:

- A high percentage of customers reorder within 1â€“2 months after their first purchase.
- Retention significantly drops after the third month, highlighting a critical window for engagement.
- Long-term customer loyalty (>6 months) is minimal without intervention strategies.

### Recommendations:
1. Implement targeted follow-up offers within 30â€“60 days of a customer's first order.
2. Create loyalty reward programs for customers who make multiple purchases within the first 90 days.
